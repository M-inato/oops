```python
# Q3  Explain the difference between instance methods and class methods. Provide an example of each.

# solution:-
'''
instance method:-

Methods that operate on the instance of a class.
They can access and modify the instance's attributes and call other instance methods.
first parameter is always the self.
these are bound to the instance of the class.

class method:-

methods that operate on the class itself rather than on an instance.
They are used for factory methods or operations that affect the whole class.
first parameter is always the cls.
these are bound to the class.
marked with the @classmethod decorator
'''

class Counter:
    count = 0

   ##this is a instance method ##
    def __init__(self):
        Counter.count += 1

   ##this is a class method##
    @classmethod
    def get_count(cls):
        return cls.count



counter1 = Counter()
counter2 = Counter()

##calling the methods
print(Counter.get_count())
```

```python
# Q7 What is the Method Resolution Order (MRO) in Python? How can you retrieve it programmatically?

# solution :-
'''
The Method Resolution Order (MRO) in Python determines the order in which classes are searched when executing a method or accessing an attribute.

we can retrieve it programmatically using the :-

Class.__mro__
Class.mro()
help(Class)

'''
class A:
    def method(self):
        print("Method in A")

class B(A):
    def method(self):
        print("Method in B")

class C(A):
    def method(self):
        print("Method in C")

class D(B, C):
    pass

# retrive MRO programatically
print(D.__mro__)
print(D.mro())
help(D)

```

```python
# Q8  Create an abstract base class `Shape` with an abstract method `area()`. Then create two subclasses `Circle` and `Rectangle` that implement the `area()` method.?

# solution :-
from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

circle = Circle(5)
rectangle = Rectangle(4, 6)

print(circle.area())
print(rectangle.area())
```

```python
# Q9 Demonstrate polymorphism by creating a function that can work with different shape objects to calculate and print their areas.

# solution:-
from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def area(self):
        return math.pi * self.radius ** 2

class Rectangle(Shape):
    def area(self):
        return self.width * self.height

def print_area(shape: Shape):
    print(shape.__class__.__name__ is shape.area())

circle = Circle()
circle.radius = 5

rectangle = Rectangle()
rectangle.width = 4
rectangle.height = 6

# Polymorphism
print_area(circle)
print_area(rectangle)
```

```python
# Q10 Implement encapsulation in a `BankAccount` class with private attributes for `balance` and `account_number`. Include methods for deposit, withdrawal, and balance inquiry.

# solution:-

class BankAccount:
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: {amount}. New Balance: {self.__balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount > 0:
            if amount <= self.__balance:
                self.__balance -= amount
                print(f"Withdrawn: {amount}. Remaining Balance: {self.__balance}")
            else:
                print("Insufficient funds.")
        else:
            print("Withdrawal amount must be positive.")

    def get_balance(self):

        return self.__balance

    def get_account_number(self):
        return self.__account_number

# Usage
account = BankAccount()

# Correctly assigning values to the private attributes
account._BankAccount__account_number = "123456789"
account._BankAccount__balance = 1000

account.deposit(500)
account.withdraw(300)
print(f"Balance: {account.get_balance()}")
print(f"Account Number: {account.get_account_number()}")
```

```python
# Q11 Write a class that overrides the `__str__` and `__add__` magic methods. What will these methods allow you to do?

# solution:-
'''
these methos allow you to :-
__str()__ :-
1.Provides a human-readable string representation of the object.

__add()__ :-
1. allow you to use the + operator  to add the two different objects.
'''
class MyNumber:
    def __init__(self, value):
        self.value = value

    def __str__(self):
        return f"MyNumber({self.value})"

    def __add__(self, other):
        if isinstance(other, MyNumber):
            return MyNumber(self.value + other.value)
        else:
            return NotImplemented

# Usage
num1 = MyNumber(10)
num2 = MyNumber(20)

# Overriding __str__
print(num1)
print(num2)

# Overriding __add__
num3 = num1 + num2
print(num3)

```

```python
# Q12 Create a decorator that measures and prints the execution time of a function ?

# solution :-
import time
def measure_time(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        execution_time = end_time - start_time
        print(f"{func.__name__}: {execution_time:} sec")
        return result
    return wrapper

@measure_time
def slow_function():
    time.sleep(2)

@measure_time
def fast_function():
    time.sleep(1)

slow_function()
fast_function()
```

```python
# Q13 Explain the concept of the Diamond Problem in multiple inheritance. How does Python resolve it?

# solution :-
'''
     A
    / \
   B   c
   \  /
    D

problem:-

if class D calls a method or try to access the data variable defined in Class A, it could be unclear whether it should call the method from Class B or Class C, since both B and C inherit from A.

This ambiguty is know as the diamond problem in the multiple inheritance.

resolving the diamond problem using the super() function.
'''
## using the super keyword to slove the diamond problem
class A:
    def speak(self):
        print("class A")

class B(A):
    def speak(self):
        print("class B")
        super().speak()

class C(A):
    def speak(self):
        print("class C")
        super().speak()

class D(B, C):
    def speak(self):
        print("class D")
        super().speak()

# Create an instance of D and call speak()
d = D()
d.speak()

```

```python
# Q15  Implement a static method in a class that checks if a given year is a leap year. ?

# solution:-
class Year:
    @staticmethod
    def isleapyear(year):
        if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
            return print("leap year")
        return print("Not leap year")

year=Year()
year.isleapyear(2024)
```
